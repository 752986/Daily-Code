```python
from matplotlib import pyplot
import numpy as np
```

```python
numlist = np.random.randint(100, size=1_000)
print(numlist)
```

```python
figure, (f1, f2, f3) = pyplot.subplots(1, 3, figsize=(20, 5))

f1.hist(numlist)
f2.hist(numlist, 10)
f3.hist(numlist, bins=[0, 25, 50, 75, 100])

pyplot.title("it work !!!", loc="center")
pyplot.show()
```
